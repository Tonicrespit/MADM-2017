### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np

from azureml import Workspace

ws = Workspace()
```

### Step 2. Import the dataset from drinks.csv

```python
ds = ws.datasets['drinks.csv']
```

### Step 3. Assign it to a variable called drinks.

```python
drinks = ds.to_dataframe()
```

### Step 4. Which continent drinks more beer on average?

```python
drinks.loc[drinks.beer_servings.idxmax()]
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks.groupby('continent')['wine_servings'].describe()
```

### Step 6. Print the mean alcoohol consumption per continent for every column

```python
drinks.groupby('continent').mean()
```

### Step 7. Print the median alcoohol consumption per continent for every column

```python
drinks.groupby('continent').median()
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
result = pd.DataFrame(columns=['mean', 'min', 'max'])
result.loc[0] = [drinks.spirit_servings.mean(), drinks.spirit_servings.min(), drinks.spirit_servings.max()]

result
```
